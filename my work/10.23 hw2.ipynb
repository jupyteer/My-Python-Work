```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x):
    return x**4
def f2(x):
    return 4*x**3
def f3(x):
    return 12*x**2
def f4(x):
    return 24*x
plt.figure(figsize = (12,6))
x = np.linspace(-10,10,num = 50)
plt.plot(x,f(x))
plt.plot(x,f2(x))
plt.plot(x,f3(x))
plt.plot(x,f4(x))
```
