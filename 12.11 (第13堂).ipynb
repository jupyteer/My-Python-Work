## Google 翻譯

可以用一個套件，叫做 googletrans

```python
!pip install googletrans
```

```python
from googletrans import Translator
```

```python
pp=Translator()
```

```python
result=pp.translate("This is a book.", dest="zh-tw")
```

```python
print(result.text)
```

```python
result2 = pp.translate("This is a book.", dest="ko")
```

```python
print(result2.text)
```

```python
#### 基本指令
'src':原文
"dest":目標語言

#常用語言
中文:'zh-tw'
日文:'ja'
韓文:'ko'
英文:'en'
```

## requests + BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.marieclaire.com.tw/love-sex/astrology"
```

```python
resp = requests.get(url)
```

```python
resp.text
```

```python
ss=BeautifulSoup(resp.text,"lxml")
```

```python
hot=ss.find_all("div",class_='hot')
```

```python
article =hot[0].find_all("a")##
```
