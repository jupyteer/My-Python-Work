## PCA  (將點投射至平面，也是個函數)

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
from sklearn.decomposition import PCA
```

 ## step 1. 開 PCA 機

```python
pca = PCA(n_components=2)
#n number                壓到2維
```

### step 2. 訓練

```python
pca.fit(x)        ###
```

### step3. 預測

```python
x[87]
```

```python
X = pca.transform(x)
```

```python
X[87]
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

### 準備訓練/ 測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train ,y_test= train_test_split(X,y,test_size=0.2,random_state=0)
```

### step 1. 開SVC函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')  #不要求太精確
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict=clf.predict(x_test)
```

```python
x0=np.arange(-4,4.3,0.02)
y0=np.arange(-1.5,1.7,0.02)

xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]

z=clf.predict(P)
Z=z.reshape(xm.shape)

plt.contourf(xm,ym,Z,alpha=0.3,cmap="Paired")
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,cmap='Paired')
```

## k-Means 自動分類

```python
x=np.random.rand(100,2)
```

```python
plt.scatter(x[:,0],x[:,1])###
```

### step 1. 打開　K-Means　學習機

```python
from sklearn.cluster import KMeans
```

```python
clf=KMeans(n_clusters=3)
```

```python
## step2
```

```python
clf.fit(x)
```

```python
##step3.
```

```python
clf.predict(x)
```

```python
clf.labels_ ##ans  因為正確答案也是電腦弄的，分類也是。
```

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_)
```

```python
x0 = y0 =np.arange(-0.2,1.2,0.02) #range same
xm,ym=np.meshgrid(x0,y0)

P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)

Z=z.reshape(xm.shape)
plt.contourf(xm,ym,Z,alpha=0.3,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired') #因為自己分的　不會有分錯
```

## Mean-shifts 完全自動分類

## step 1打開函數學習機

```python
from sklearn.cluster import MeanShift
```

```python
clf=MeanShift()  #can use bandwidth in() to adjust N groups
```

## step2

```python
clf.fit(x)
```

### step3.預測

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

```python
x0=np.arange(-4.2,5.2,0.02)
y0=np.arange(-3,7,0.02)

xm,ym=np.meshgrid(x0,y0)

P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)

Z=z.reshape(xm.shape)

plt.contourf(xm,ym,Z,alpha=0.3,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_)
```
