```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
```

```python
boston_dataset = load_boston()
```

```python
boston = pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
boston['MEDV'] = boston_dataset.target
```

```python
x = boston.loc[:,"RM":"LSTAT"].values
y = boston.MEDV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
plt.style.use("ggplot")
```

```python
fig = plt.figure(figsize=(6,4*13)) 
for i in range(0,13): 
    plt.subplot2grid((13,1),(i,0)) 
    plt.scatter(boston_dataset.data[:,i],boston_dataset.target,c='blue') 
    plt.xlabel(boston_dataset.feature_names[i]) 
    plt.ylabel('Price') 
    plt.show
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print ('The value of R-squared of LinearRegression is', r2_score(y_test, y_predict)
```
